```python
%matplotlib inline

from datetime import datetime

import matplotlib.pyplot as plt
import numpy as np

plt.title('Hot Water')
plt.xlabel('Time(s) since Epoch')
plt.ylabel('Celsius')

# when disabling the xticks, the grid has only horizontal lines.
# we still have yticks, so I doubt the usability of grid altogether.
# plt.grid()

# one hour, 5s interval
x = np.arange(0, 3600, 5)

plt.plot(x, (60-30)/3600 * x + 30, label='Boiler(top)')
plt.plot(x, (50-25)/3600 * x + 25, label='Boiler(middle)')
plt.plot(x, (40-20)/3600 * x + 20, label='Boiler(middle)')

plt.legend()
#plt.tight_layout()
```
