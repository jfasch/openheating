```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
```

# Spline Interpolation

```python
x = [1, 1.9, 2.5, 2.6, 3.3, 5]
y = [1.3, 2.2, 2.3, 2.321, 1.2, 8]
plt.plot(x,y)
```

```python
y_splfunc = interpolate.splrep(x,y)
```

```python
y_splfunc
```

Aha. Apparently an iternal representation of the interpolated curve.
Generate new (and more) points, smoothing the curve.

```python
x = np.arange(1,5,0.1)
y = interpolate.splev(x, y_splfunc)
plt.plot(x,y)
```

Interpolation outside the interval

```python
x = np.arange(0,6,0.1)
y = interpolate.splev(x, y_splfunc)
plt.plot(x,y)
```
